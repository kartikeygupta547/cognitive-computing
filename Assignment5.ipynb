Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
Find

i. Sum of all elements

ii. Sum of all elements row-wise

iii. Sum of all elements column-wise


```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
total_sum = np.sum(gfg)
print("Sum of all element in a matrix",total_sum)
row_sum = np.sum(gfg, axis=1)
print("sum of all element row wise")
print(row_sum)
column_sum = np.sum(gfg, axis=0)
print("Sum of all element column wise",column_sum)
```

Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]),
 find

i. Sorted array

ii. Indices of sorted array

iii. 4 smallest elements

iv. 5 largest elements

```python
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted array",sorted_array)
sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)
smallest_elements = sorted_array[:4]
print("4 smallest elements:", smallest_elements)
largest_elements = sorted_array[-5:]
print("5 largest elements:", largest_elements)
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]),
find

i. Integer elements only

ii. Float elements only


```python
import numpy as np
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array[array == array.astype(int)]
print("Integer elements only:", integer_elements)
float_elements = array[array != array.astype(int)]
print("Float elements only:", float_elements)
```

Q.3 You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.

a) Generate your unique sales dataset:
 Take the sum of the ASCII values of the initials of your first and last
name. Call this value X.

(If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)

 Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].


b) Compute your personalized tax rate as ((X % 5) + 5) / 100.

 Use broadcasting to apply this tax rate to each sales value.

c) Adjust sales based on discount:

 If sales < X+100, apply a 5% discount.

 If sales >= X+100, apply a 10% discount.

d) Expand sales data for multiple weeks:

 Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.

 Increase sales by 2% per week using element-wise broadcasting.

```python

import numpy as np
initials = "GG"
X = sum(ord(c) for c in initials)
sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Sales Dataset:", sales)
tax_rate = ((X % 5) + 5) / 100
sales_with_tax = sales * (1 + tax_rate)
print("Sales with Tax:", sales_with_tax)
discount_rates = np.array([0.05, 0.10])
sales_with_discount = np.where(sales < X + 100, sales * (1 - discount_rates[0]), sales * (1 - discount_rates[1]))
print("Sales with Discount:", sales_with_discount)
sales = np.array([131, 181, 231, 281, 331])
weekly_sales = np.tile(sales, (3, 1))
weekly_sales = weekly_sales * (1 + 0.02)**np.arange(3)[:, np.newaxis]
print(weekly_sales)
```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:

 Y = x2

 Y = sin(x)

 Y = ex

 Y = log(|x| + 1)

Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)
chosen_function = y1
plt.plot(x, chosen_function)
plt.title('Plot of Chosen Function (Y = x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
print("\n")
chosen_function=y2
plt.plot(x, chosen_function)
plt.title('Plot of Chosen Function (Y = sin(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
print("\n")
chosen_function=y3
plt.plot(x, chosen_function)
plt.title('Plot of Chosen Function (Y = exp(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
print("\n")
chosen_function=y4
plt.plot(x, chosen_function)
plt.title('Plot of Chosen Function (Y = log(|x|+1))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
